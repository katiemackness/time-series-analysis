In time series, a **random walk** is a process in which there is an equal chance of going up or down by a random number. <br> <br>
In this notebook, I work through some examples and complete some practice exercises as a means of understanding and practicing random walk processes. Examples come from the book *Time Series Forecasting in Python* by Marco Peixeiro.

## Simulate a Random Walk Process

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# Set the random seed
np.random.seed(42) 
# Generate 1,000 random numbers from a normal distribution
steps = np.random.standard_normal(1000)
```

```python
# Initialize the first value of the series to 0
steps[0] = 0
# Calculate the cumulative sum of errors
random_walk = np.cumsum(steps)
```

```python
# Plot the simulated random walk
fig, ax = plt.subplots()

ax.plot(random_walk)
ax.set_xlabel('Timesteps')
ax.set_ylabel('Value')

plt.tight_layout
```

## Identifying a Random Walk
To be a true random walk, the points in time must be stationary and uncorrelated. 

With **stationarity**, the statistical properties of a time series do not change over time. It maintains a constant mean, variance and autocorrelation. Since many forecasting models assume stationarity, this characteristic must be verified first in order to build valid models upon the data.
