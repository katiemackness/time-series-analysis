```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
apple = yf.download(tickers='aapl', period = '2y', interval = '1mo', rounding=True)
```

```python
apple
```

```python
apple = apple['Close']
apple
```

```python
apple.index = pd.to_datetime(apple.index)
apple2024 = apple.loc['2024']
```

```python
apple2024.plot(grid=True)
```

```python
rendimientos = apple.pct_change()
rendimientos
```

```python
apple_microsoft  = yf.download(tickers='aapl msft', period='2y', interval='1mo', rounding=True)
apple_microsoft
```

```python
apple_microsoft = apple_microsoft['Close']
```

```python
apple_microsoft['AAPL'].plot()
apple_microsoft['MSFT'].plot()
plt.legend(['Apple', 'Microsoft'])
plt.title('Comparación de precios Apple - Microsoft')
plt.show()
```

```python
rend_apple_microsoft = apple_microsoft.pct_change()
```
