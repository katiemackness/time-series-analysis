```python
# Prueba de caminata aleatoria
# H0: Se tiene un proceso de caminata aleatoria
# Ha: No se tiene un proceso de caminata aleatoria

from statsmodels.tsa.stattools import adfuller
import numpy as np

# Generacin de 500 pasos aleatorios
np.random.seed(1)
steps = np.random.normal(loc=0.001, scale=0.01, size=500)+1
```

```python
steps[0] = 1
P = 100 * np.cumprod(steps)
```

```python
plt.plot(P)
plt.title("Proceso de Caminata Aleatoria Pura")
plt.show()
```

```python
resultado = adfuller(P)
print("El valor p de la prueba es:" + str(resultado[1]))
```

```python
# Prueba Dickey-Fuller para los precios de APPLE
resultado = adfuller(apple)
print("El valor p de la prueba es:" + str(resultado[1]))
```

```python
# Definicin de grupos de entrenamiento y prueba
train = apple[0:len_train]
test = apple[len_train:]
```

```python
y_pred = apple.copy()
y_pred = pd.DataFrame(y_pred)
```

```python
y_pred['pronostico'] = apple.rolling(30).mean()
y_pred[0:35]
```

```python
y_pred.tail()
```

```python
y_pred_m = y_pred.copy()
y_pred_m['pronostico'][29] = 'NaN'
for contador in range(30, len(y_pred_m)):
    y_pred_m['pronostico'][contador] = y_pred['pronostico'][contador - 1]

y_pred_m[0:35]
```

```python
y_pred_m = pd.DataFrame(y_pred_m['pronostico'][604:])
y_pred_m
```

```python
plt.figure(figsize=(20,5))
plt.grid()
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(y_pred_m['pronostico'], label="Prediccin", color='red')
plt.legend(loc='best')
plt.title("Pronsticos mediante medias mviles")
plt.show()
```

```python
test = pd.DataFrame(test)
test
```

```python
frames = [test, y_pred_m]
result = pd.concat(frames, axis=1, join='inner')
result.columns = ['Observado', 'Pronosticado']
result
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse = np.sqrt(mean_squared_error(test, y_pred_m)).round(2)
mape = np.round(np.mean(np.abs((test['AAPL'] - y_pred_m['pronostico']) / test['AAPL'])) * 100, 2)
```

```python
print("RMSE =", rmse, "\nMAPE =", mape)
```
