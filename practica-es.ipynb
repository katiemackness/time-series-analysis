```python
# Prueba de caminata aleatoria
# H0: Se tiene un proceso de caminata aleatoria
# Ha: No se tiene un proceso de caminata aleatoria

from statsmodels.tsa.stattools import adfuller
import numpy as np

# Generación de 500 pasos aleatorios
np.random.seed(1)
steps = np.random.normal(loc=0.001, scale=0.01, size=500)+1
```

```python
steps[0] = 1
P = 100 * np.cumprod(steps)
```

```python
plt.plot(P)
plt.title("Proceso de Caminata Aleatoria Pura")
plt.show()
```

```python
resultado = adfuller(P)
print("El valor p de la prueba es:" + str(resultado[1]))
```

```python
# Prueba Dickey-Fuller para los precios de APPLE
resultado = adfuller(apple)
print("El valor p de la prueba es:" + str(resultado[1]))
```

```python
# Graficación de la media móvil para Apple
rolling_apple = apple.rolling(30).mean()

plt.figure()
plt.plot(apple, label="Apple")
plt.plot(rolling_apple, label='Media móvil')
plt.title("Precios históricos de Apple vs. Promedios Móviles a 30 días")
plt.legend()
plt.show()
```
